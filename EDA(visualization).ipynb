# EDA

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("data.csv")
```

```python
df.head(20)
```

```python
df.info()
```

```python
df.describe()
```

```python
# Create age ranges
df['Age_Range'] = (df['Age'] // 10) * 10 + 1

# Define age range bins
age_bins = [13, 20, 30, 40, 50, 56]

# Map credit score categories to numerical values
credit_score_mapping = {'Good': 1, 'Standard': 2, 'Poor': 3}
df['Credit_Score_Category'] = df['Credit_Score'].map(credit_score_mapping)

# Create a new column for credit score categories based on conditions
df['Credit_Score_Category'] = pd.cut(df['Credit_Score_Category'], bins=3, labels=['Good', 'Standard', 'Poor'])

# Create a cross-tabulation of Age_Range and Credit_Score_Category
cross_tab = pd.crosstab(df['Age_Range'], df['Credit_Score_Category'])

# Plot the cross-tabulated data
cross_tab.plot(kind='bar', stacked=True, colormap='viridis', figsize=(10, 6))

# Set plot labels and title
plt.xlabel('Age Range')
plt.ylabel('Count')
plt.title('Age Range vs. Credit Score')

# Show the plot
plt.show()
```

```python

# Create a cross-tabulation of Occupation and Credit_Score
cross_tab_occupation = pd.crosstab(df['Occupation'], df['Credit_Score'])

# Define custom colors for each credit score category
colors = {'Good': 'green', 'Standard': 'orange', 'Poor': 'red'}

# Plot the cross-tabulated data with custom colors
ax = cross_tab_occupation.plot(kind='bar', stacked=True, color=[colors[col] for col in cross_tab_occupation.columns], figsize=(14, 8))

# Set plot labels and title
plt.xlabel('Occupation')
plt.ylabel('Count')
plt.title('Occupation vs. Credit Score')

# Annotate bars with counts
for p in ax.patches:
    width, height = p.get_width(), p.get_height()
    x, y = p.get_xy()
    ax.text(x + width/2, y + height/2, f'{height}', ha='center', va='center', fontsize=8, color='white')

# Show the plot
plt.show()

```

## What we can know
    1. There are no big difference between occpupations.

```python
# Plot the data
plt.figure(figsize=(12, 6))
ax = sns.countplot(x='Num_Bank_Accounts', hue='Credit_Score', data=df, palette={'Good': 'green', 'Standard': 'orange', 'Poor': 'red'})

# Set plot labels and title
plt.xlabel('Number of Bank Accounts')
plt.ylabel('Count')
plt.title('Number of Bank Accounts vs. Credit Score')

# Show the plot
plt.show()
```

## What we can know
    1. As the number of bank accounts increases, the credit score decreases.
    2. There is a sharp increase in the people in poor category once the number of bank accounts goes over 5.
    3.good category sees a sharp decrease after 5 bank accounts.

```python
# Scatter plot
plt.figure(figsize=(10, 6))

# Define colors for credit score categories
colors = {'Good': 'green', 'Standard': 'orange', 'Poor': 'red'}

# Scatter plot for each credit score category
for credit_score_category, color in colors.items():
    subset = df[df['Credit_Score'] == credit_score_category]
    plt.scatter(subset['Num_Credit_Card'], subset['Credit_Score'], label=credit_score_category, color=color, alpha=0.5)

# Set plot labels and title
plt.xlabel('Number of Credit Cards')
plt.ylabel('Credit Score')
plt.title('Number of Credit Cards vs. Credit Score')

# Add legend
plt.legend()

# Show the plot
plt.show()
```

## What we can know
    1. As the number of credit cards increases, the credit score decreases.
    2. When the number of credit cards is 3 or 7-8, the credit score drops significantly.

```python
# Scatter plot
plt.figure(figsize=(10, 6))

# Define colors for credit score categories
colors = {'Good': 'green', 'Standard': 'orange', 'Poor': 'red'}

# Scatter plot for each credit score category
for credit_score_category, color in colors.items():
    subset = df[df['Credit_Score'] == credit_score_category]
    plt.scatter(subset['Interest_Rate'], subset['Credit_Score'], label=credit_score_category, color=color, alpha=0.5)

# Set plot labels and title
plt.xlabel('Interest Rate')
plt.ylabel('Credit Score')
plt.title('Interest Rate vs. Credit Score')

# Add legend
plt.legend()

# Show the plot
plt.show()
```

## What we can know

```python
# Set the style for the plot
sns.set(style="whitegrid")

# Create a bar plot
plt.figure(figsize=(10, 6))
sns.countplot(x='Payment_of_Min_Amount', hue='Credit_Score', data=df, palette='viridis')

# Set plot labels and title
plt.xlabel('Payment of Minimum Amount')
plt.ylabel('Count')
plt.title('Payment of Minimum Amount vs. Credit Score')

# Show the plot
plt.show()
```

```python
# Calculate normalized counts
normalized_counts = df.groupby(['Payment_Behaviour', 'Credit_Score']).size().reset_index(name='Percentage')
normalized_counts['Percentage'] /= normalized_counts.groupby('Payment_Behaviour')['Percentage'].transform('sum') * 0.01

# Set the style for the plot
sns.set(style="whitegrid")

# Create a normalized stacked bar plot
plt.figure(figsize=(10, 6))
sns.barplot(x='Payment_Behaviour', y='Percentage', hue='Credit_Score', data=normalized_counts, palette='viridis')

# Set plot labels and title
plt.xlabel('Payment Behaviour')
plt.ylabel('Percentage')
plt.title('Payment Behaviour vs. Credit Score')

# Rotate x-axis labels for better visibility
plt.xticks(rotation=45)

# Show the plot
plt.show()
```

## What we can know
    1. The number of people in the Low_spent_Medium_value_payments group is the highest. But their credit scores are the lowest.
    2. People in the High_spent_Large_value_payments group have the best credit scores.

```python
# Set the style for the plot
sns.set(style="whitegrid")

# Creating a histogram
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='Annual_Income', hue='Credit_Score', multiple='stack', bins=30, palette='viridis')

# Set plot labels and title
plt.xlabel('Annual Income')
plt.ylabel('Count')
plt.title('Annual Income Distribution by Credit Score')

# Show the plot
plt.show()
```
